# IMPORTANDO A BASE DE DADOS

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.width', 400)         # MOSTRAR TODAS AS COLUNAS HORIZONTALMENTE
pd.set_option('display.max_columns', None)  # NÃO HÁ LIMITE DE COLUNAS PRA MOSTRAR
pd.set_option('display.max_rows', 10)       # MOSTRA SEMPRE 10 LINHAS
```

```python
df = pd.read_excel(r'DataBase.xlsx', sheet_name='Clientes')
df.head(5)
```

# REMOVENDO COLUNAS DESNECESSÁRIAS
Vamos remover todas as colunas que contém a palavra "unnamed" em seu nome e também a coluna ID cliente, que é desnecessária

```python
def findColumns(df, nameRemove):
    lista = []
    for coluna in df.columns:
        if nameRemove.lower() in coluna.lower():
            lista.append(coluna)
    return lista
```

```python
colunasInuteis = findColumns(df, 'unnamed')

df = df.drop(columns=colunasInuteis)
df = df.drop(columns=['IDCliente'])

df.head()
```

# TRATAMENTO DE ERROS
Colocando como NaN todas as linhas que possuem um valor em branco (só com espaços)

```python
def invalidarEspacos(valor):
    if str(valor).strip() == '':
        return np.nan
    else:
        return valor
    
df = df.applymap(invalidarEspacos)
```

Removendo todas as colunas completamente nulas (só com NaN)

```python
df = df.dropna(how='all', axis=1)
```

Removendo todas as linhas que possuem algum valor NaN

```python
df = df.dropna(how='any', axis=0)
```

Tabela formatada sem valores NaN

```python
df.head(6)
```

# TRATAMENTO DOS VALORES NUMÉRICOS

```python
df.info()
```

Vemos que existem colunas de valores numéricos como "TotalGasto" que estão como objeto, ou seja, há alguma string neles, para arrumar isso vamos converter essa coluna para numérico. Depois preenchemos os valores com erro pelo valor 0

```python
df['TotalGasto'] = pd.to_numeric(df['TotalGasto'], errors='coerce')
df['TotalGasto'] = df['TotalGasto'].fillna(0)

df.head(6)
```

# TRATAMENTO DOS VALORES DATA
A coluna "Mês de Assinatura" deve também ser convertida para o formato de data. Depois preenchemos os valores com erro pelo valor 0

```python
df['Mês de Assinatura'] = pd.to_datetime(df['Mês de Assinatura'], errors='coerce')
df['TotalGasto'] = df['TotalGasto'].fillna(0)

df.head(6)
```

# SALVANDO A TABELA FORMATADA

```python
df.to_csv(r'CleanDataBase.csv', index=False)
```
