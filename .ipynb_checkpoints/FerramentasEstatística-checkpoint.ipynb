# AMPLITUDE
É a diferença entre o maior e menor valores de uma base de dados

```python
def Amplitude(vetor=[]):
    return max(vetor) - min(vetor)
```

```python
lista = [2, 2, 4, 5, 7, 8, 9, 10]
Amplitude(lista)
```

# MÉDIA
Soma dos valores dividido pelo número de valores

```python
def media(vetor=[]):
    soma = 0
    tamanho = len(vetor)
    for c in vetor:
        soma += c
    return soma / tamanho
```

```python
lista = [2, 2, 4, 5, 7, 8, 9, 10]
media(lista)
```

# MEDIANA
É o valor que está no meio do vetor, se o tamanho do vetor for um número par, é são a média dos dois valores do meio

```python
def mediana(vetor=[]):
    resto = len(vetor) % 2        # DESCOBRINDO SE O RESTO É PAR OU ÍMPAR
    vetor.sort()                  # ORDENANDO O VETOR
    
    if resto == 0:                # SE FOR PAR
        meio = len(vetor) // 2    # PEGA SÓ A PARTE INTEIRA ARREDONDADA PRA BAIXO
        return (vetor[meio - 1] + vetor[meio]) / 2
    else:
        meio = len(vetor) / 2
        return vetor[int(meio)]   # AUTOMATICAMENTE ARREDONDA PRA BAIXO
```

```python
lista = [2, 2, 4, 5, 7, 8, 9, 10]
mediana(lista)
```

# MODA
É a quantidade que mais se repete na base de dados

```python
def moda(vetor=[]):
    semRepetidas = []
    for c in range(len(vetor)):             # CRIANDO UMA LISTA SEM VALORES REPETIDOS
        if vetor[c] not in semRepetidas:
            semRepetidas.append(vetor[c])
            
    vetorContagem = []
    for c in range(len(semRepetidas)):      # CONTANDO OS VETORES DESSA LISTA
        contagem = vetor.count(semRepetidas[c])
        vetorContagem.append(contagem)

    maiorContagem = max(vetorContagem)
    posicao = vetorContagem.index(maiorContagem) # POSIÇÃO 

    return semRepetidas[posicao]
```

```python
lista = [0, 2, 4, 5, 7, 8, 7, 7, 7, 10]
moda(lista)
```

# MÉDIA PONDERADA

```python
def mediaPonderada(valores, pesos):
    soma = 0
    somaPeso = 0
    for c in range(0, len(valores)):
        soma += valores[c] * pesos[c]
        somaPeso += pesos[c]
    return soma / somaPeso
```

```python
notas = [5, 7, 7, 5, 12, 5]
peso =  [0.1, 0.1, 0.1, 0.3, 0.2, 0.2]

mediaPonderada(notas, peso)
```

# DESVIO PADRÃO E VARIÂNCIA
$\sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{n}}$, onde $\sigma$ é o desvio padrão e $\mu$ é a média

```python
def desvioPadrao(vetor=[]):
    soma = 0
    for c in vetor:
        soma += (c - media(vetor))**2
    variancia = soma / len(vetor)
    desvio = (variancia)**(1/2)
    return desvio
```

```python
lista = [0, 2, 4, 5, 7, 8, 7, 7, 7, 10]
desvioPadrao(lista)
```

Desvio amostral $\rightarrow \sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{n- 1}}$

```python
def desvioAmostral(vetor=[]):
    soma = 0
    for c in vetor:
        soma += (c - media(vetor)) ** 2
    variancia = soma / (len(vetor) - 1)
    desvio = (variancia)**(1/2)
    return desvio
```

```python
lista = [0, 2, 4, 5, 7, 8, 7, 7, 7, 10]
desvioAmostral(lista)
```

# COEFICIENTE DE VARIAÇÃO
Descreve o desvio padrão como uma porcentagem da média

```python
def coeficienteVariacao(vetor, amostral=False):
    if amostral:
        return desvioAmostral(vetor) / media(vetor) * 100
    else:
        return desvioPadrao(vetor) / media(vetor) * 100
```

```python
lista = [0, 2, 4, 5, 7, 8, 7, 7, 7, 10]
coeficienteVariacao(lista)
```

# OUTLIERS
Valores que se afastam muito da média (usar fórmula).

```python
def isOutlier(dados, x):
    tabela = [0, 0, 0, 1.38, 1.54, 1.65, 1.73, 1.80, 1.85, 1.91, 1.96, 1.99, 2.03, 2.06,
              2.1, 2.13, 2.16, 2.18, 2.20, 2.22, 2.24, 2.26, 2.28, 2.30, 2.31, 2.33,
              2.35, 2.36, 2.37, 2.38, 2.39]
    
    n = len(dados)
    r = abs(media(dados) - x)/desvioPadrao(dados)
    R = tabela[n]
    
    if r > R:
        return True
    
    return False

def allOutliers(dados):
    outliers = []
    for x in dados:
        if isOutlier(dados, x):
            outliers.append(x)

    return outliers
```

```python
lista = '2,477	2,472	2,471	2,446	2,460'

lista = lista.replace(',', '.').replace('	', ',')
lista = lista.split(',')
for c in range(len(lista)):
    lista[c] = float(lista[c])

print(lista)
print('OUTLIERS:', allOutliers(lista))
```

# FREQUÊNCIA ACUMULATIVA
Lista com as contagens de cada valor (frequência)

```python
# CRIANDO LISTA SEM DUPLICATAS
def semRepetidas(vetor=[]):
    vetor2 = []
    for c in vetor:
        if c not in vetor2:
            vetor2.append(c)
    return vetor2

def vetorFrequencia(vetor=[]):
    vetor_unico = semRepetidas(vetor)
    frequencia = []
    for c in range(0, len(vetor_unico)):
        contador = vetor.count(vetor_unico[c])
        frequencia.append( contador )
    return frequencia
```

```python
lista = [-50, 2, 2, 4, 5, 5, 7, 200, 8, 7, 7, 7, 10, 150]
vetorFrequencia(lista)
```

# FREQUÊNCIA RELATIVA
É a porcentagem de cada valor (contagem) da frequência acumulativa em relação ao total

```python
lista = [-50, 2, 2, 4, 5, 5, 7, 200, 8, 7, 7, 7, 10, 150]
frequencias = vetorFrequencia(lista)

frequenciaRelativa = [c/sum(frequencias) for c in frequencias]
frequenciaRelativa
```

# FREQUÊNCIA ACUMULATIVA COM CLASSES
Dividimos o vetor em um número $x$ de classes e contamos sua frequência em um determinado intervalo de cada classe

```python
# LISTA COM OS INTERVALOS ENTRE O MENOR E MAIOR VALORES
def limites(lista, nClasses):
    amplitude = Amplitude(lista)
    classe = int(amplitude / nClasses) if (amplitude % nClasses == 0) else int(amplitude / nClasses + 1)
    intervalo = []
    soma = min(lista)

    intervalo.append(min(lista))
    while soma < max(lista):
        soma += classe
        intervalo.append(soma)
    return intervalo

def frequenciaAcumulativaClasses(vetor=[], nClasses=0):
    frequencia = limites(vetor, nClasses)
    vetorFreq = [0 for c in frequencia]

    for c in range(0, len(frequencia)-1):
        for v in range(0, len(vetor)):
            if frequencia[c] <= vetor[v] < frequencia[c+1]:
                vetorFreq[c] += 1
    return vetorFreq[:-1] # O ÚLTIMO NÃO CONTA PQ É A FREQUÊNCIA ENTRE DOIS VALORES
```

```python
lista = [-50, 2, 2, 4, 5, 5, 7, 200, 8, 7, 7, 7, 10, 150, 20, 10]
limites(lista, 5)
```

```python
frequenciaAcumulativaClasses(lista, 5)
```

# FREQUENCIA RELATIVA COM CLASSES
Mostra a porcentagem dos valores em relação ao total em cada intervalo das classes

```python
def frequenciaRelativaClasses(vetor=[], nClasses=0):
    vetorFreq = frequenciaAcumulativaClasses(vetor, nClasses)
    vetorFreqRel = []
    for c in range(0, len(vetorFreq)):
        freq = vetorFreq[c]/sum(vetorFreq)
        freq = round(freq, 2)
        vetorFreqRel.append(freq)
    return vetorFreqRel
```

```python
lista = [-50, 2, 2, 4, 5, 5, 7, 200, 8, 7, 7, 7, 10, 150, 20, 10]
frequenciaRelativaClasses(lista, 5)
```

# LIMITES COM PONTO MÉDIO
Mostra o ponto médio dos intervalos das classes

```python
def limitesPontoMedio(lista, nClasses):
    vetorLimite = limites(lista, nClasses)
    vetorMedio = []
    for c in range(0, len(vetorLimite)-1):
        vetorMedio.append( (vetorLimite[c] + vetorLimite[c+1]) / 2)
    return vetorMedio
```

```python
lista = [-50, 2, 2, 4, 5, 5, 7, 200, 8, 7, 7, 7, 10, 150, 20, 10]
limitesPontoMedio(lista, 5)
```

# INCERTEZA DA MÉDIA
Usando a tabela T-Student para encontrar a incerteza da média de um conjunto de valores distibuídos normalmente (suas frequências se concentram em torno da média)

```python
def incertezaMedia(lista):
    n = len(lista)
    tStudent = [
        12.706, 4.303, 3.182, 2.776, 2.571, 2.447, 2.265, 2.306, 2.262, 2.228, 2.201, 2.179,
        2.160, 2.145, 2.131, 2.120, 2.110, 2.101, 2.093, 2.086, 2.080, 2.074, 2.069, 2.064,
        2.060, 2.056, 2.052, 2.048, 2.045, 2.042
    ]

    if n <= 30:
        return tStudent[n-2] * desvioAmostral(lista)/(n**0.5)
    else:
        return desvioAmostral(lista)/(n**0.5)
```

```python
lista = [-50, 2, 2, 4, 5, 5, 7, 200, 8, 7, 7, 7, 10, 150, 20, 10]
incertezaMedia(lista)
```

# ELIMINAÇÃO DE OUTLIERS
Método bastante utilizado para identificar valores discrepantes que atrapalham em uma análise e conclusão geral dos dados

```python
def isOutlier(dados, x):
    tabela = [0, 0, 0, 1.38, 1.54, 1.65, 1.73, 1.80, 1.85, 1.91, 1.96, 1.99, 2.03, 2.06,
              2.1, 2.13, 2.16, 2.18, 2.20, 2.22, 2.24, 2.26, 2.28, 2.30, 2.31, 2.33,
              2.35, 2.36, 2.37, 2.38, 2.39]
    
    n = len(dados)
    r = abs(media(dados) - x)/desvioPadrao(dados)
    R = tabela[n]
    
    if r > R:
        return True
    
    return False

def allOutliers(dados):
    outliers = []
    for x in dados:
        if isOutlier(dados, x):
            outliers.append(x)

    return outliers

def removeValues(dados, lista):
    for x in lista:
        for y in range(dados.count(x)):
            dados.remove(x)
        
    return dados
```

```python
lista = [2.459, 2.462, 2.465, 2.467, 2.453]

lista = removeValues(lista, allOutliers(lista))
lista
```
